```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from sklearn import linear_model
import seaborn as sns
```

```python
lineal = pd.read_csv("C:\date\lineal\data.csv")
```

```python
lineal.head(5)
```

```python
lineal.plot.scatter(x="metro",y="precio")
plt.show()
```

```python
regresion = linear_model.LinearRegression()
```

```python
metros = lineal["metro"].values.reshape((-1,1))
```

```python
modelo = regresion.fit(metros, lineal["precio"])
```

```python
print("Interseccion (b)",modelo.intercept_)
```

```python
print("Pendiente (m)",modelo.coef_)
```

```python
entrada = [[500],[375],[60],[20]]
predicciones = modelo.predict(entrada)
print(predicciones)
```

```python
lineal.plot.scatter(x="metro",y="precio", label='Datos orginales')
plt.scatter(entrada,predicciones,color='red')
plt.plot(entrada,predicciones, color='black',label='linea de regresion')
plt.xlabel('metro')
plt.ylabel('precio')
plt.legend()
plt.show()
```
